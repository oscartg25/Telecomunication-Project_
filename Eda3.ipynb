```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tipos_accesos = pd.read_csv('Dataset/tipos_accesos_telefonomovil.csv', encoding='utf-8')
```

```python
df_tipos_accesos
```

```python
# Verificar los tipos de datos
print(df_tipos_accesos.dtypes)

# Tratar los valores nulos
print(df_tipos_accesos.isnull().sum())
```

```python
# Limpiar formatos de datos (ejemplo: eliminar puntos en los números)
df_tipos_accesos['Total de accesos pospago'] = df_tipos_accesos['Total de accesos pospago'].str.replace('.', '').astype(int)
df_tipos_accesos['Total de accesos prepago'] = df_tipos_accesos['Total de accesos prepago'].str.replace('.', '').astype(int)
df_tipos_accesos['Total de accesos operativos'] = df_tipos_accesos['Total de accesos operativos'].str.replace('.', '').astype(int)

# Eliminar columnas innecesarias
#df_tipos_accesos = df_tipos_accesos.drop(['Periodo'], axis=1)
```

```python
df_tipos_accesos
```

```python
# Gráfico de líneas
df_tipos_accesos.plot(x='Año', y=['Total de accesos pospago', 'Total de accesos prepago'], kind='line')
plt.xlabel('Año')
plt.ylabel('Total de accesos')
plt.title('Tendencia de accesos pospago y prepago')
plt.legend()
plt.show()
```

# TELEFONIA FIJA

```python
import datetime
```

```python
df_telefonia_fija = pd.read_csv('Dataset/telefonia_fija.csv', encoding='utf-8')
```

```python
df_telefonia_fija
```

```python
df_telefonia_fija['Accesos telefonía fija por cada 100 hab'] = df_telefonia_fija['Accesos telefonía fija por cada 100 hab'].str.replace(',', '.').str.strip()
```

```python
# Convertir la columna a tipo numérico
df_telefonia_fija['Accesos telefonía fija por cada 100 hab'] = pd.to_numeric(df_telefonia_fija['Accesos telefonía fija por cada 100 hab'])
```

```python
# Convertir la columna a tipo numérico
df_telefonia_fija['Accesos telefonía fija por cada 100 hogares'] = df_telefonia_fija['Accesos telefonía fija por cada 100 hogares'].str.replace(',', '.').str.strip()
```

```python
df_telefonia_fija['Accesos telefonía fija por cada 100 hogares'] = pd.to_numeric(df_telefonia_fija['Accesos telefonía fija por cada 100 hogares'])
```

```python
df_telefonia_fija
```

```python
# Histograma
df_telefonia_fija['Accesos telefonía fija por cada 100 hab'].plot(kind='hist', bins=10)
plt.xlabel('Accesos telefonía fija por cada 100 hab')
plt.ylabel('Frecuencia')
plt.title('Distribución de accesos telefonía fija por cada 100 hab')
plt.show()
```

```python
promedios_por_provincia_y_año = df_telefonia_fija.groupby(['Provincia', 'Año'])['Accesos telefonía fija por cada 100 hogares'].mean()
print(promedios_por_provincia_y_año)
```

```python
# promedios_por_provincia_y_año = df_telefonia_fija.pivot_table(index='Año', columns='Provincia', values='Accesos telefonía fija por cada 100 hogares', aggfunc='mean')
```

```python
# Obtener el año actual
año_actual = datetime.datetime.now().year

# Filtrar los últimos tres años
df_ultimos_tres_años = df_telefonia_fija[df_telefonia_fija['Año'] >= año_actual - 2]

# Crear la tabla pivot con los últimos tres años
promedios_por_provincia_y_año = df_ultimos_tres_años.pivot_table(index='Año', columns='Provincia', values='Accesos telefonía fija por cada 100 hogares', aggfunc='mean')
```

```python
# Gráfico de barras
promedios_por_provincia_y_año.plot(kind='bar')
plt.xlabel('Año')
plt.ylabel('Promedio de accesos por cada 100 hogares')
plt.title('Promedio de accesos telefonía fija por cada 100 hogares por provincia en los últimos 2 años')
plt.legend(title='Provincia', bbox_to_anchor=(1, 1))
plt.show()
```
